### cost estmation 

```python
import csv
from datetime import datetime


def cost_calculation(response) -> float:
    """
    Calculate API cost based on token usage.
    """

    prompt_tokens = response.usage.input_tokens
    completion_tokens = response.usage.output_tokens

    # GPT-4.1 mini pricing
    input_cost = 0.4 / 1000000
    output_cost = 1.6 / 1000000

    cost = (prompt_tokens * input_cost) + \
           (completion_tokens * output_cost)

    return cost


def create_csv_if_not_exists(csv_file: str):
    """
    Create CSV file with headers if it does not exist or is empty.
    """

    with open(csv_file, mode="a", newline="") as file:
        writer = csv.writer(file)

        if file.tell() == 0:
            writer.writerow([
                "timestamp",
                "prompt_tokens",
                "completion_tokens",
                "total_tokens",
                "cost_usd"
            ])


def log_cost_to_csv(response, csv_file: str = "api_cost_log.csv"):
    """
    Log API usage and cost to CSV file.
    """

    cost = cost_calculation(response)

    row = [
        datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        response.usage.input_tokens,
        response.usage.output_tokens,
        response.usage.total_tokens,
        f"{cost:.6f}"
    ]

    create_csv_if_not_exists(csv_file)

    with open(csv_file, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(row)

    return cost




cost = log_cost_to_csv(response)
print(f"Estimated cost of the API call: ${cost:.6f}")
```

### Lesson 7 — Add Conversation Memory / History

```python

#load libaries and set up API client
from dotenv import load_dotenv
from openai import OpenAI
import os

# function to require environment variable and raise error if not found
def require_key(name: str) -> str:
    value = os.getenv(name)
    if not value:
        raise RuntimeError(f"Missing environment variable: {name}")
    return value

#load environment variables from .env file and create OpenAI client
load_dotenv()
OPENAI_MODEL = "gpt-4.1-mini"
client = OpenAI(api_key=require_key("OPENAI_API_KEY"))


#define function to ask question to the AI and get response

def ask_ai_with_history(
    conversation_history: list[dict[str, str]],
    instructions: str = "You are a helpful AI assistant.",
    temperature: float = 0.7,
    max_output_tokens: int = 400,) -> str:

    response = client.responses.create(
        model=OPENAI_MODEL,
        instructions=instructions,
        input=conversation_history,
        temperature=temperature,
        max_output_tokens=max_output_tokens,

    )

    return response.output_text

# Example usage: ask a question about RAG and fine-tuning

def run_chatbot() -> None:
    print("AI Chatbot started.")

    print("Type 'exit' to stop.\n")
    instructions = "You are a beginner-friendly AI engineering lecturer. Explain clearly and simply."
    conversation_history = []

    while True:
        user_input = input("You: ").strip()

        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        if not user_input:
            print("Please type a question.")
            continue

        # adding user input and AI response to conversation history for better context in future interactions
        conversation_history.append({
            "role": "user",
            "content": user_input,
        })

        answer = ask_ai_with_history(
            conversation_history=conversation_history, 
            # conversation_history=conversation_history[-10:],  # Optionally limit history to last 10 messages for context
            instructions=instructions,
            temperature=0.3,
            max_output_tokens=400,
        )

        #adding AI response to conversation history
        conversation_history.append({
            "role": "assistant",
            "content": answer,
        })

        print(f"User: {user_input}\n")
        print(f"AI: {answer}\n")
    
    print("Conversation history:", conversation_history)

if __name__ == "__main__":
    run_chatbot()
```

### Lesson 8 — Using previous_response_id for Conversation Memory

```python
from dotenv import load_dotenv
from openai import OpenAI
import os


def require_key(name: str) -> str:
    value = os.getenv(name)

    if not value:
        raise RuntimeError(f"Missing environment variable: {name}")

    return value


load_dotenv()

OPENAI_MODEL = "gpt-4.1-mini"

client = OpenAI(api_key=require_key("OPENAI_API_KEY"))


def ask_ai_with_previous_response_id(
    user_input: str,
    previous_response_id: str | None = None,
    instructions: str = "You are a helpful AI assistant.",
    temperature: float = 0.3,
    max_output_tokens: int = 400,
):
    response = client.responses.create(
        model=OPENAI_MODEL,
        instructions=instructions,
        input=user_input,
        previous_response_id=previous_response_id,
        temperature=temperature,
        max_output_tokens=max_output_tokens,
    )

    return response


def run_chatbot() -> None:
    print("AI Chatbot with previous_response_id started.")
    print("Type 'exit' to stop.\n")

    instructions = "You are a beginner-friendly AI engineering lecturer. Explain clearly and simply."

    previous_response_id = None

    while True:
        user_input = input("You: ").strip()

        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        if not user_input:
            print("Please type a question.")
            continue

        response = ask_ai_with_previous_response_id(
            user_input=user_input,
            previous_response_id=previous_response_id,
            instructions=instructions,
            temperature=0.3,
            max_output_tokens=400,
        )

        previous_response_id = response.id

        print(f"AI: {response.output_text}\n")


if __name__ == "__main__":
    run_chatbot()
```

```python
from dotenv import load_dotenv
from openai import (
    OpenAI,
    APIConnectionError,
    APITimeoutError,
    APIStatusError,
    AuthenticationError,
    RateLimitError,
)
import logging
import os


logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(name)s | %(message)s",
)

logger = logging.getLogger(__name__)


def require_key(name: str) -> str:
    value = os.getenv(name)

    if not value:
        logger.error("Missing required environment variable: %s", name)
        raise RuntimeError(f"Missing environment variable: {name}")

    return value


load_dotenv()

OPENAI_MODEL = "gpt-4.1-mini"

client = OpenAI(api_key=require_key("OPENAI_API_KEY"))


def ask_ai(
    user_input: str,
    instructions: str = "You are a helpful AI assistant.",
    temperature: float = 0.3,
    max_output_tokens: int = 300,
) -> str | None:
    logger.info("Sending request to OpenAI model=%s", OPENAI_MODEL)

    try:
        response = client.responses.create(
            model=OPENAI_MODEL,
            instructions=instructions,
            input=user_input,
            temperature=temperature,
            max_output_tokens=max_output_tokens,
        )

        logger.info("OpenAI response received successfully")
        return response.output_text

    except AuthenticationError:
        logger.exception("Authentication failed")
        return "Authentication failed. Please check your OPENAI_API_KEY."

    except RateLimitError:
        logger.exception("Rate limit reached")
        return "Rate limit reached. Please wait and try again."

    except APITimeoutError:
        logger.exception("OpenAI request timed out")
        return "The request timed out. Please try again."

    except APIConnectionError:
        logger.exception("Network connection problem")
        return "Network connection problem. Please check your internet connection."

    except APIStatusError as error:
        logger.exception("OpenAI API status error: %s", error.status_code)
        return f"OpenAI API error: status={error.status_code}, message={error.message}"

    except Exception:
        logger.exception("Unexpected error")
        return "Unexpected error occurred."


if __name__ == "__main__":
    logger.info("Lesson 10 app started")

    answer = ask_ai(
        user_input="Explain RAG in one sentence.",
        instructions="You are a beginner-friendly AI engineering lecturer.",
    )

    print(answer)

    logger.info("Lesson 10 app finished")
```
